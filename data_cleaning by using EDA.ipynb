```python
dataFile['Age'] = dataFile['Age'].astype(str).str.replace(r'[^0-9]','',regex=True)
dataFile
```

```python
dataFile['Age'] = dataFile['Age'].str.extract(r'(\d+)')
dataFile
```

```python
dataFile['Exp'] = dataFile['Exp'].astype(str).str.extract(r'(\d+)')
dataFile
```

```python
dataFile['Location'] = dataFile['Location'].str.replace(r'\W','',regex=True)
dataFile
```

```python
dataFile['Salary'] = dataFile['Salary'].astype(str).str.replace(r'\W','',regex=True)
dataFile
```

```python
clean_data['Age']=pd.to_numeric(clean_data['Age'])
clean_data['Age']=clean_data['Age'].fillna(clean_data['Age'].mean())
clean_data['Age']
```
